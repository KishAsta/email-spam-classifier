Import necessary libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from google.colab import drive
# Mount Google Drive
drive.mount('/content/drive')
```

Load dataset from CSV file into a pandas DataFrame

```python
df = pd.read_csv('/content/drive/My Drive/mail.csv', encoding='latin1')
```

Display the loaded DataFrame

```python
print(df)
```

Replace missing values with empty strings

```python
data=df.where((pd.notnull(df)),'')
```

Display the first 10 rows of the cleaned data

```python
data.head(10)
```

Display information about the dataset, including data types and non-null counts

```python
data.info()
```

```python
data.shape
```

Convert 'spam' and 'ham' labels to numeric values (0 for 'spam', 1 for 'ham')

```python
data.loc[data['Category']=='spam','Category',]=0
data.loc[data['Category']=='ham','Category',]=1
```

Extract features (X) and labels (Y)

```python
X=data['Message']
Y=data['Category']
```

Display features and labels

```python
print(X)
```

```python
print(Y)
```

Split the dataset into training and testing sets

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=3)
```

Display the shapes of the datasets

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
print(Y.shape)
print(Y_train.shape)
print(Y_test.shape)
```

Initialize and configure TfidfVectorizer for feature extraction

```python
feature_extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)
# Transform the text data into TF-IDF features
X_train_features=feature_extraction.fit_transform(X_train)
X_test_features=feature_extraction.transform(X_test)
# Convert label data to integers
Y_train=Y_train.astype('int')
Y_test=Y_test.astype('int')
```

 Display the transformed training data

```python
print(X_train)
```

```python
print(X_train_features)
```

Initialize and train a Logistic Regression model

```python
model=LogisticRegression()
```

```python
model.fit(X_train_features,Y_train)
```

Make predictions on the training sets



```python
prediction_on_training_data=model.predict(X_train_features)
accuracy_on_training_data=accuracy_score(Y_train,prediction_on_training_data)
```

Display accuracy on the training set

```python
print('Accuracy on training data:',accuracy_on_training_data)
```

Make predictions on the test sets



```python
prediction_on_test_data=model.predict(X_test_features)
accuracy_on_test_data=accuracy_score(Y_test,prediction_on_test_data)
```

Display accuracy on the test set

```python
print('Accuracy on test data:',accuracy_on_test_data)
```
